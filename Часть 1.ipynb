Взяв случайные данные из таблицы можно сказать, что среднее колчество спален примерно равно 0,25 количества комнат. 

```python
nan_indexes = housing_train_dummies[housing_train_dummies['average_bedrooms'].isnull()].index
fill_data = housing_train_dummies['average_rooms']*0.25
housing_train_dummies.loc[nan_indexes, 'average_bedrooms'] = pd.Series(fill_data, index=nan_indexes)
```

```python
longitude_array = np.array(housing_train_dummies['longitude'])
latitude_array = np.array(housing_train_dummies['latitude'])

normalized_longitude = preprocessing.normalize([longitude_array])
normalized_latitude = preprocessing.normalize([latitude_array])

housing_train_dummies.loc[:,'longitude'] = normalized_longitude[0]
housing_train_dummies.loc[:,'latitude'] = normalized_latitude[0]
```

```python
def get_dummies_for_test(train_columns, test_df, dummy_sep):
    initial_df = pd.get_dummies(test_df, prefix_sep=dummy_sep)
    missing_in_test = set(train_columns) - set(initial_df.columns)
    
    for c in missing_in_test:
        initial_df[c] = 0
    return initial_df.loc[:,train_columns].copy()
```

```python
housing_test_dummies = get_dummies_for_test(housing_train_dummies.columns, housing_train_dummies, DUMMY_SEP)
housing_val_dummies = get_dummies_for_test(housing_train_dummies.columns, housing_train_dummies, DUMMY_SEP)
```
