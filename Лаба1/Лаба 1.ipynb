```python
with open("sales.json", "r") as sales:
    data = json.load(sales)
titles = ['item', 'country', 'year', 'sales']
with open('sales.csv', 'w') as f: 
    wr = csv.DictWriter(f, fieldnames = titles) 
    wr.writeheader() 
    for item in data:
        countrys = item["sales_by_country"]
        for country in countrys:
            years = countrys[country]
            for year in years: 
                wr.writerow({'item':item['item'],'country':country,'year':year,'sales':item['sales_by_country'][country][year]})
```

```python
def function(X):
    X1 = X.copy()
    averageX = np.mean(X1, axis=0)
    stdX = np.std(X1, axis=0)
    X1 = (X1 - averageX)/stdX
    return X1


d = random.randrange(2,6,1)
N = random.randrange(2,100,1)
X = np.random.uniform(1,100,size=(N,d))
X1 = function(X)
newX = np.random.multivariate_normal(mean=[1, 2], cov=[[2, 1], [1, 3]],size=50)
newX1 = function(newX)

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(121)

s=0
m=0
if d == 2:
    for m in range (newX.shape[0]):    
        dot1 = plt.scatter(newX[m][0],newX[m][1], color='#2b82d9')
        dot2 = ax.scatter(newX1[m][0],newX1[m][1], color='#00bfff')
    for s in range (X.shape[0]):
        dot3 = plt.scatter(X[s][0],X[s][1],  color='#ad655f')
        dot4 = ax.scatter(X1[s][0],X1[s][1],  color='#fc0fc0')
    plt.legend([dot1,dot2,dot3,dot4],['изначальные точки первой матрицы','выходные точки первой матрицы','изначальные точки второй матрицы','выходные точки второй матрицы'])
plt.show()
```
