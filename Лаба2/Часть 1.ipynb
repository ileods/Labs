```python
housing_df = pd.read_csv('housing.csv')
```

```python
housing_df
```

```python
DUMMY_SEP = '__'
housing_df_dummies = pd.get_dummies(housing_df, prefix_sep=DUMMY_SEP, columns=['ocean_proximity'])
```

```python
housing_df_dummies
```

```python
dummy_columns = [c for c in housing_df_dummies.columns if DUMMY_SEP in c]
```

```python
housing_df_dummies
```

```python
housing_train_val, housing_test = train_test_split(housing_df_dummies, test_size=0.1)
housing_train, housing_val = train_test_split(housing_train_val, test_size = 0.1)
```

```python
housing_train
```

```python
housing_train.insert(3,'average_rooms',housing_train.loc[:,'total_rooms']/housing_train.loc[:,'households'])
housing_train.insert(4,'average_bedrooms',housing_train.loc[:,'total_bedrooms']/housing_train.loc[:,'households'])
housing_train = housing_train.drop(['total_rooms','total_bedrooms'], axis=1)
```

```python
housing_train
```

```python
housing_test.insert(3,'average_rooms',housing_test.loc[:,'total_rooms']/housing_test.loc[:,'households'])
housing_test.insert(4,'average_bedrooms',housing_test.loc[:,'total_bedrooms']/housing_test.loc[:,'households'])
housing_test = housing_test.drop(['total_rooms','total_bedrooms'], axis=1)
```

```python
housing_test
```

```python
housing_val.insert(3,'average_rooms',housing_val.loc[:,'total_rooms']/housing_val.loc[:,'households'])
housing_val.insert(4,'average_bedrooms',housing_val.loc[:,'total_bedrooms']/housing_val.loc[:,'households'])
housing_val = housing_val.drop(['total_rooms','total_bedrooms'], axis=1)
```
