```python
housing_df = pd.read_csv('housing.csv')
```

```python
housing_df
```

```python
DUMMY_SEP = '__'
housing_df_dummies = pd.get_dummies(housing_df, prefix_sep=DUMMY_SEP, columns=['ocean_proximity'])
```

```python
housing_df_dummies
```

```python
dummy_columns = [c for c in housing_df_dummies.columns if DUMMY_SEP in c]
dummy_columns_to_remove = [c for c in dummy_columns if housing_df_dummies[c].sum() < 10] 
```

```python
dummy_columns_to_remove
```

```python
housing_df_dummies = housing_df_dummies.drop(dummy_columns_to_remove, axis=1)
```

```python
housing_df_dummies
```

```python
housing_train_val, housing_test = train_test_split(housing_df_dummies, test_size=0.1)
housing_train, housing_val = train_test_split(housing_train_val, test_size = 0.1)
```

```python
housing_train
```

```python
housing_train.insert(3,'average_rooms',housing_train.loc[:,'total_rooms']/housing_train.loc[:,'households'])
housing_train.insert(4,'average_bedrooms',housing_train.loc[:,'total_bedrooms']/housing_train.loc[:,'households'])
housing_train = housing_train.drop(['total_rooms','total_bedrooms'], axis=1)
```

```python
housing_train
```

```python
sum(housing_train['average_bedrooms'].isnull())
```

Для 166 элементов отсутсвует признак average_bedrooms.

```python
housing_train.sample(10)
```

Взяв случайные данные из таблицы можно сказать, что среднее колчество спален примерно равно 0,25 количества комнат. 

```python
nan_indexes = housing_train[housing_train['average_bedrooms'].isnull()].index
fill_data = housing_train['average_rooms']*0.25
housing_train.loc[nan_indexes, 'average_bedrooms'] = pd.Series(fill_data, index=nan_indexes)
```

```python
sum(housing_train['average_bedrooms'].isnull())
```

```python
housing_train['longitude'] = preprocessing.scale(housing_train['longitude'])
housing_train['latitude'] = preprocessing.scale(housing_train['latitude'])
```
